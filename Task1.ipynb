### 1.1 Создать новый признак price_per_sq_lot, который будет содержать стоимость за один кв. метр общей площади

````python
df['price_per_sq_lot'] =  df['price'] / df['sqft_lot']
````

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

Создадим lambda-функцию с проверкой заданного условия  

````python
df['delta_renovated'] = df.apply(lambda x: x['yr_renovated'] - x['yr_built'] if x['yr_renovated'] != 0 else 0, axis=1)
# Функция apply() используется для применения lambda-функции к каждой строке DataFrame. 
# Параметр axis=1 указывает, что функция должна применяться к каждой строке, а не к каждому столбцу/
````

### 1.3 Создайте признаки года продажи, месяца продажи

Используем метод `apply()`, который применяет lambda-функцию к каждому элементу столбца **date**. </br> lambda-функция `x[:4]` делает срез и извлекает из строки даты первые 4 сивола, то есть год.</br> lambda-функция `x[4:6]` делает срез и извлекает из строки даты символы с 4 по 6, то есть месяц.

````python
df['Year_Sale'] = df['date'].apply(lambda x: x[:4])
df['Month_Sale'] = df['date'].apply(lambda x: x[4:6])
````

проверим наличие новых признаков

````python
df.info()
````

### 1.4 Удалите признаки date, zipcode, lat, long

Для удаления используем метод `.drop()`

````python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True) 
  #  аргумент inplace=True указывает на то, что изменения внесены в сам DataFrame, а не создан новый.
````

Проверим факт удаления
